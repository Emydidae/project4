```python
# changing numbers to the categorical data they represent
def categories(column, values):
    for i in range(len(column)):
        column[i] = values[column[i] - 1]
    return(column)
```

```python
df = pd.read_csv('raw_data.csv')
df
```

```python
new_names = [
    'STUDENT ID', 'Student Age', 'Sex', 'High-school type', 'Scholarship type', 'Additional work',
    'Art/Sports', 'Marital Status', 'Salary', 'Transportation to the university',
    'Accommodation type in Cyprus', "Mother's education", "Father's education",
    'Sisters/brothers', 'Parental Marital Status', "Mother's occupation",
    "Father's occupation", 'Weekly study hours', 'Reading frequency (non-science)',
    'Reading frequency (science)', 'Seminars/Conferences', 'Projects/Activities',
    'Attendance to classes', 'Preparation to midterm exams 1', 'Preparation to midterm exams 2',
    'Taking notes in classes', 'Listening in classes', 'Discussion', 'Flip-classroom',
    'GPA (/4.00)', 'Exp GPA', 'COURSE ID', 'GRADE'
]
df.columns = new_names
df.set_index('STUDENT ID', inplace = True)
df.head()
```

```python
# update data
categorical_df = df.copy()
for i in range(len(categorical_df.columns)):
    col = categorical_df.columns[i]
    categorical_df[col] = categories(categorical_df[col].to_list(), cats[i])
categorical_df.head()
```
